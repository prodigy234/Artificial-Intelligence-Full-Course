# 1. Core Python Visualization Libraries

## a) Matplotlib

- Low-level, very flexible

- Foundation of most Python visualization libraries

- Best for learning fundamentals

```python
import matplotlib.pyplot as plt
```

## b) Seaborn

- High-level, statistical visualizations

- Built on Matplotlib

- Excellent for data analysis & teaching patterns

```python
import seaborn as sns
```

## c) Plotly

- Interactive, web-friendly

- Great for dashboards and storytelling

```python
import plotly.express as px
```

## 2. Line Chart

When to Use

- Time series data

- Trends over time

- Continuous data

Example Use-Cases

- Stock prices over time

- Monthly sales growth

- Temperature changes

```python
# Sample Data
days = [1, 2, 3, 4, 5]
sales = [200, 250, 300, 280, 350]

plt.plot(days, sales)
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Daily Sales Trend')
plt.show()
```

## 3. Bar Chart

When to Use

- Comparing categories

- Discrete data

Example Use-Cases

- Revenue by product

- Students per department

- Votes per candidate

```python
products = ['A', 'B', 'C']
revenue = [5000, 7000, 6000]

plt.bar(products, revenue)
plt.title('Revenue by Product')
plt.show()
```

## 4. Histogram

When to Use

- Distribution of a single variable

- Frequency analysis

Example Use-Cases

- Exam scores distribution

- Age distribution

- Income distribution

```python
scores = [45, 50, 55, 60, 65, 70, 75, 80, 85]
plt.hist(scores, bins=5)
plt.title('Score Distribution')
plt.show()
```

## 5. Scatter Plot

When to Use

- Relationship between two numerical variables

- Correlation detection

Example Use-Cases

- Height vs Weight

- Advertising spend vs sales

```python
hours = [1,2,3,4,5]
scores = [50,55,65,70,80]
plt.scatter(hours, scores)
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.title('Study Hours vs Scores')
plt.show()
```

## 6. Box Plot

When to Use

- Identify outliers

- Compare distributions

Example Use-Cases

- Salary comparison across departments

- Exam scores by class

```python
scores = [50,55,65,70,80]
sns.boxplot(data=[scores])
plt.title('Score Spread')
plt.show()
```

## 7. Violin Plot

When to Use

- Distribution + density

- More informative than box plot

Example Use-Cases

- Customer spending patterns

- Exam score distributions

```python
scores = [50,55,65,70,80]
sns.violinplot(data=scores)
plt.title('Score Distribution Density')
plt.show()
```

## 8. Pie Chart

When to Use

- Part-to-whole relationship

- Small number of categories

Example Use-Cases

- Market share

- Budget allocation

```python
items = ['Food', 'Rent', 'Transport']
sizes = [40, 35, 25]


plt.pie(sizes, labels=items, autopct='%1.1f%%')
plt.title('Monthly Expenses')
plt.show()
```

## 9. Area Chart

When to Use

- Cumulative trends

- Time-based comparisons

Example Use-Cases

- Revenue growth

- Population growth

```python
days = [1, 2, 3, 4, 5]
sales = [200, 250, 300, 280, 350]
plt.fill_between(days, sales)
plt.title('Sales Area Chart')
plt.show()
```

## 10. Heatmap

When to Use

- Correlation analysis

- Pattern detection in matrices

Example Use-Cases

- Feature correlation

- Student performance matrix

```python
import numpy as np

data = np.random.rand(5,5)
sns.heatmap(data, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

## 11. Pair Plot

When to Use

- Multivariate analysis

- Exploratory Data Analysis (EDA)

Example Use-Cases

- Feature relationships in ML

```python
sns.pairplot(sns.load_dataset('iris'))
plt.show()
```

## 12. Count Plot

When to Use

- Frequency of categorical data

Example Use-Cases

- Gender count

- Product category frequency

```python
data = ['Male','Female','Male','Female','Male']
sns.countplot(x=data)
plt.show()
```

## 13. Bubble Chart

When to Use

- 3-variable relationship

Example Use-Cases

- GDP vs Population vs Life Expectancy

```python
hours = [1,2,3,4,5]
scores = [50,55,65,70,80]
plt.scatter(hours, scores, s=[50,100,200,300,400])
plt.show()
```

## 14. Interactive Charts (Plotly Example)

```python
df = px.data.iris()
fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species')
fig.show()
```

## Summary

Goal	                    Best Chart

Trend           ==>	        Line Chart

Comparison      ==>	        Bar Chart

Distribution    ==>	        Histogram / Box

Relationship    ==>	        Scatter

Correlation     ==>	        Heatmap

Composition     ==>	        Pie / Area
