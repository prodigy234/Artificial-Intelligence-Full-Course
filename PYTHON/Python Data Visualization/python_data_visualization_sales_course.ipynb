## 3. Data Preparation for Visualization

```python

df['sale_date'] = pd.to_datetime(df['sale_date'])

df['month'] = df['sale_date'].dt.month
df['year'] = df['sale_date'].dt.year

df.head()
```

`   df['sale_date'] = pd.to_datetime(df['sale_date'])   `

Explanation:

This converts the sale_date column into datetime format.

Why?

Because datetime allows:

- Time analysis

- Month extraction

- Year extraction

- Time series charts

Before conversion:

**"2023-03-01"**

After conversion:

**Timestamp('2023-03-01')**

`   df['month'] = df['sale_date'].dt.month  `

Explanation:

This extracts the month from the date.

Example:

sale_date	month

2023-03-01	3

2023-04-10	4

Why?

To analyze:

- Monthly revenue

- Monthly sales trends

`   df['year'] = df['sale_date'].dt.year    `

Extracts the year.

Useful for:

- Yearly trends

- Multi-year datasets

# MATPLOTLIB VISUALIZATIONS

## 4. Line Chart (Sales Trend Over Time)

```python

sales_trend = df.groupby('sale_date')['total_price'].sum()

plt.figure(figsize=(10,5))
plt.plot(sales_trend.index, sales_trend.values, marker='o')
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Revenue")
plt.show()
```

`   sales_trend = df.groupby('sale_date')['total_price'].sum()  `

Explanation:

- This groups data by date.

`   groupby('sale_date')    `

- Then calculates total revenue per date.

`   plt.figure(figsize=(10,5))  `

**Creates a figure canvas.**

figsize controls size.

- 10 = width

- 5 = height

`   plt.plot(sales_trend.index, sales_trend.values, marker='o') `

**Creates a line chart.**

Components:

x-axis

`   sales_trend.index   `

**This shows the Date on the x-axis of the line chart.**

y-axis

`   sales_trend.values  `

**This shows the Revenue values on the y-axis of the line chart.**

`   marker='o'  `

Adds dots to each data point.

**Add title**

`   plt.title("Sales Trend Over Time")  `

- Sets chart title.

**Axis labels**

`   plt.xlabel("Date")  `

`   plt.ylabel("Total Revenue") `

- Helps readers understand the chart.

**Show chart**

`   plt.show()  `

- Displays the plot.

**Without show(), the chart may not render.**

### Practical Insight
Line charts are best for **time series data**.

## 5. Bar Chart (Revenue by Product)

```python

product_sales = df.groupby('product_id')['total_price'].sum()

plt.figure(figsize=(8,5))
plt.bar(product_sales.index, product_sales.values)
plt.title("Revenue by Product")
plt.xlabel("Product ID")
plt.ylabel("Revenue")
plt.show()
```

## 6. Histogram (Distribution of Sales Revenue)

```python

plt.figure(figsize=(8,5))
plt.hist(df['total_price'], bins=10)
plt.title("Distribution of Revenue")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

Histograms help analysts **understand the distribution of numerical data**.

## 7. Scatter Plot (Quantity vs Revenue)

```python

plt.figure(figsize=(8,5))
plt.scatter(df['quantity_sold'], df['total_price'])
plt.xlabel("Quantity Sold")
plt.ylabel("Revenue")
plt.title("Quantity vs Revenue")
plt.show()
```

Scatter plots help identify **relationships between variables**.

# SEABORN VISUALIZATIONS

## 8. Barplot with Seaborn

```python

sns.barplot(x='product_id', y='total_price', data=df)

plt.title("Average Revenue by Product")
plt.show()
```

```python
ax = sns.barplot(x='product_id', y='total_price', data=df)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}',
                 (p.get_x() + p.get_width() / 2. , p.get_height()), 
                 ha='center', va='bottom')

plt.title("Average Revenue by Product")
plt.show()
```

## 9. Box Plot (Outlier Detection)

```python

sns.boxplot(x=df['total_price'])

plt.title("Revenue Outlier Detection")
plt.show()
```

Boxplots help identify **outliers and spread of data**.

## 10. Violin Plot

```python

sns.violinplot(x=df['product_id'], y=df['total_price'])

plt.title("Revenue Distribution by Product")
plt.show()
```

Violin plots combine **boxplot + density distribution**.

## 11. Count Plot (Sales Frequency by Product)

```python

sns.countplot(x='product_id', data=df)

plt.title("Sales Count by Product")
plt.show()
```

## 12. Heatmap (Correlation Matrix)

```python

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)

plt.title("Correlation Matrix")
plt.show()
```

Heatmaps help visualize **relationships between numerical variables**.

## 13. Pairplot (Multivariate Analysis)

```python

sns.pairplot(df)
plt.show()
```

Pairplots help analysts **see relationships between many variables simultaneously**.

# TIME SERIES VISUALIZATION

## 14. Monthly Revenue Trend

```python

monthly_sales = df.groupby('month')['total_price'].sum()

plt.figure(figsize=(8,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

# ADVANCED VISUALIZATION

## 15. Subplots (Multiple Charts in One Figure)

```python

fig, ax = plt.subplots(1,2, figsize=(12,5))

product_sales = df.groupby('product_id')['total_price'].sum()
ax[0].bar(product_sales.index, product_sales.values)
ax[0].set_title("Revenue by Product")

ax[1].hist(df['total_price'])
ax[1].set_title("Revenue Distribution")

plt.show()
```
