## 3. Loading Data from CSV

```python
# Example loading CSV
# df = pd.read_csv("sales_data.csv")

# For demonstration we generate a dataset
np.random.seed(42)

sales = pd.DataFrame({
    "order_id": range(1,101),
    "region": np.random.choice(["North","South","East","West"],100),
    "sales_rep": np.random.choice(["Alice","Bob","Charles","Diana"],100),
    "product": np.random.choice(["Laptop","Phone","Tablet","Monitor"],100),
    "quantity": np.random.randint(1,10,100),
    "price": np.random.randint(200,2000,100)
})

sales["revenue"] = sales["quantity"] * sales["price"]

sales.head()
```

## 4. Inspecting Data

```python
sales.head()
sales.tail()
sales.sample(5)
```

```python
sales.shape
sales.columns
sales.info()
```

```python
sales.describe()
```

## 5. Selecting Columns

```python
sales["revenue"]
```

```python
sales[["product","revenue"]].head()
```

## 6. Filtering Rows

```python
# revenue greater than 5000
sales[sales["revenue"] > 5000].head()
```

```python
# multiple conditions
sales[(sales["region"]=="North") & (sales["revenue"]>5000)].head()
```

## 7. Sorting Data

```python
sales.sort_values("revenue", ascending=False).head()
```

## 8. Creating New Columns (Feature Engineering)

```python
sales["discount"] = np.random.uniform(0.05,0.20,len(sales))

sales["net_revenue"] = sales["revenue"] * (1 - sales["discount"])

sales.head()
```

## 9. Handling Missing Data

```python
# introduce missing values
sales.loc[0:5,"sales_rep"] = np.nan

sales.isna().sum()
```

```python
# drop missing
sales.dropna().head()
```

```python
# fill missing
sales["sales_rep"] = sales["sales_rep"].fillna("Unknown")

sales.head()
```

## 10. GroupBy Analysis

```python
sales.groupby("region")["revenue"].sum()
```

```python
sales.groupby("product")["revenue"].mean()
```

```python
sales.groupby(["region","product"])["revenue"].sum()
```

## 11. Aggregation Functions

```python
sales.groupby("region").agg({
    "revenue":["sum","mean","max"],
    "quantity":"sum"
})
```

## 12. Pivot Tables

```python
pd.pivot_table(
    sales,
    values="revenue",
    index="region",
    columns="product",
    aggfunc="sum"
)
```

## 13. Merging DataFrames

```python
customers = pd.DataFrame({
    "sales_rep":["Alice","Bob","Charles","Diana"],
    "experience":[5,8,3,10]
})

merged = pd.merge(sales, customers, on="sales_rep", how="left")

merged.head()
```

## 14. Concatenation

```python
df1 = sales.head(5)
df2 = sales.tail(5)

pd.concat([df1, df2])
```

## 15. Working with Datetime

```python
dates = pd.date_range(start="2024-01-01", periods=100)

sales["date"] = dates

sales["month"] = sales["date"].dt.month

sales.head()
```

## 16. Time Series Analysis

```python
sales.groupby("month")["revenue"].sum()
```
