# Pandas DataFrames

## What is a DataFrame?

A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
# Example

# Create a simple Pandas DataFrame:

import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df) 
```

Why df.loc[0] is considered a single row

In pandas, a row is defined by its index label, not by how many values you see.

Your DataFrame looks like this conceptually:

   calories  duration
0       420        50

1       380        40

2       390        45


Row 0 = everything with index label 0

That row has two columns, so you see two values

But it is still one row, because there is one index label

#### What df.loc[0] actually returns

calories    420

duration     50

Name: 0, dtype: int64

#### This means:

0 → the row index

calories, duration → column names

Values → the data in that one row

#### So pandas collapses the row into a Series:

Index = column names

Values = row values

#### Think of it like this

Row = horizontal record (index-based)

Column = vertical attribute

```python
# Return row 0 and 1:

import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

#use a list of indexes:
print(df.loc[[0, 1]])      # Note: When using [], the result is a Pandas DataFrame.
```
