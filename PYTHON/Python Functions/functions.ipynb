Python Functions

A function is a block of code which only runs when it is called.

You can pass data, known as parameters, into a function.

A function can return data as a result.

Creating a Function

In Python a function is defined using the def keyword:

Defining a Function

In Python, you define a function using the def keyword.

Syntax:

def function_name(parameters):

    """Optional docstring explaining the function."""
    
    # block of code

This is a very well detailed explanation on the different components in the syntax above when creating a function

* function_name — the name of your function (follow standard naming conventions).

* parameters — inputs (optional) the function needs to work with.

* """Docstring""" — (optional) a brief description of what the function does.

* Indented block — code that the function executes.

```python
# e.g

def my_function():
  print("Hello from a function")
```

Calling a Function

After defining it, you call a function by writing its name followed by parentheses ().

In other words, to successfully call a function, use the function name followed by parenthesis:

```python
# e.g

def my_function():
  """I created this function to say HI from the function."""
  print("Hello from a function")

my_function()
```

```python
# another example

def greet():
    print("Hello, welcome to Python!")

greet()  # Function call
```

```python
def greet_user(name):
    print(f"Hello, {name}! Welcome to Python....")

greet_user("Segun")
```

Parameters and Arguments

Parameter: A variable listed inside the parentheses in the function definition.

Argument: A value sent to the function when it is called.

Arguments

Information can be passed into functions as arguments.

Arguments are specified after the function name, inside the parentheses. You can add as many arguments as you want, just separate them with a comma.

The example below has a function with one argument (first_name). When the function is called, we pass along a first name, which is used inside the function to print the full name:

```python
def my_function(first_name):
  print(" Hi " + first_name)

my_function("Comfort")
my_function("Tata")
my_function("Ajanlekoko")
```

```python
my_function("Gbenga")
```

```python
def greet_user(name):  # 'name' is the parameter
    print(f"Thank you {name} for investing with us!")

greet_user("Comfort")  # 'Comfort' is the argument
```

You can have multiple parameters, separated by commas.

```python
def multiply_numbers(a, b):
    return a * b

result = multiply_numbers(15, 30)
print(result)
```

Number of Arguments

By default, a function must be called with the correct number of arguments. Meaning that if your function expects 2 arguments, you have to call the function with 2 arguments, not more, and not less.

```python
# In the example below, this function expects 2 arguments, and gets 2 arguments:

def my_function(fname, lname):
  print(fname + " " + lname)

my_function("Comfort", "Ajanlekoko")
```

NOTE => if you try to call the function created above with 1 or 3 arguments, you will get an error as shown in the example below. 
This is because while creating the function, two arguments was created with it.
So for error not to be elicited while calling the function, the number of parameters created must match the number of arguments when it is called.

```python
# This function expects 2 arguments, but gets only 1:

def my_function(fname, lname):
  print(fname + " " + lname)

my_function("Emil")
```

```python
# BAR CHARTS WITHOUT USING FUNCTIONS

# Importing of packages
import matplotlib.pyplot as plt

# My data
students = ["Ade", "Bola", "Chidi", "Zainab"]
scores = [78, 85, 69, 90]

# My plot
plt.bar(students, scores)

# Labels and Titles
plt.title("Student Scores")
plt.xlabel("Students")
plt.ylabel("Scores")

# Show chart
plt.show()
```

```python
# Importing of packages
import matplotlib.pyplot as plt

# My data
students = ["Ade", "Bola", "Chidi", "Zainab"]
scores = [78, 85, 69, 90]

def Gbefun(students, scores, title, xlabel, ylabel):
    plt.bar(students, scores)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

Gbefun(students, scores, "Student Scores", "Students", "Scores")
```
