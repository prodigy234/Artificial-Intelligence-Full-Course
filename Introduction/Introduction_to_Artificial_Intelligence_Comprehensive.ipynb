## 1. What is Artificial Intelligence?

Artificial Intelligence (AI) refers to the simulation of human intelligence in machines.
It enables systems to perform tasks that typically require human intelligence such as:
- Learning
- Reasoning
- Problem-solving
- Perception
- Language understanding

### Key Goals of AI:
- Create intelligent agents
- Automate decision-making
- Enhance human capabilities


## 2. History of AI

- 1950: Alan Turing proposes the Turing Test
- 1956: Dartmouth Conference (Birth of AI)
- 1980s: Expert Systems Boom
- 2010s: Deep Learning Revolution
- 2020s: Generative AI Explosion

## 5. Machine Learning Overview

Machine Learning (ML) is a subset of AI that allows systems to learn from data.

### Types of ML:
- Supervised Learning
- Unsupervised Learning
- Semi-supervised Learning
- Reinforcement Learning

### Common Algorithms:
- Linear Regression
- Logistic Regression
- Decision Trees
- Random Forest
- K-Means


## 6. Deep Learning

Deep Learning is a subset of ML using Neural Networks.

### Concepts:
- Artificial Neural Networks (ANN)
- Activation Functions
- Backpropagation
- Convolutional Neural Networks (CNN)
- Recurrent Neural Networks (RNN)
- Transformers
